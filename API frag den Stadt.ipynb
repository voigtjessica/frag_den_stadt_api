So now I'm going to do a for loop in which I always check if there is a value for 'next' , and if it is true I add 50 to the final number of the offset.
I'm doing this way because I want to use this notebook in the future, and if the database increases I won't be able to get it with the same code. 
The original solution I created here: https://github.com/Transparencia-Brasil/API_TDP_Python

```python
def fetch_api(url):
    a = list()
    while url != None :
        r = requests.get(url).json()
        b = r['objects']
        a = a + b
        url = r['meta']['next']
        print(url)
    return a
```

Let's run!

```python
url = 'https://fragdenstaat.de/api/v1/request/'

objects = fetch_api(url)
```
